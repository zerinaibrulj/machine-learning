```python
from scipy.stats import skew, kurtosis

def _norm_col(name):
    return ''.join(ch.lower() for ch in str(name) if ch.isalnum())

desired_blue_columns = [
    'Flight Distance',
    'Departure Delay',
    'Arrival Delay',
    'Departure and Arrival Time Convenience',
    'Ease of Online Booking',
    'Check-in Service',
    'Online Boarding',
    'Gate Location',
    'On-board Service',
    'Seat Comfort',
    'Leg Room Service',
    'Cleanliness',
    'Food and Drink',
    'In-flight Service',
    'In-flight Wifi Service',
    'In-flight Entertainment',
    'Baggage Handling',
]

norm_to_actual = {_norm_col(c): c for c in df.columns}
resolved_columns = [
    norm_to_actual[_norm_col(col)]
    for col in desired_blue_columns
    if _norm_col(col) in norm_to_actual
]
missing_columns = [
    col for col in desired_blue_columns
    if _norm_col(col) not in norm_to_actual
]

rows = []
for col in resolved_columns:
    s = pd.to_numeric(df[col], errors='coerce')
    s_clean = s.dropna()
    
    zero_count = int((s == 0).sum())
    mean_with_zero = s.mean()
    mean_without_zero = s[s != 0].mean()
    mean_change = mean_without_zero - mean_with_zero
    mean_change_pct = (mean_change / mean_with_zero * 100) if pd.notna(mean_with_zero) and mean_with_zero != 0 else np.nan
    
    # New statistics
    variance = s_clean.var()
    std_dev = s_clean.std()
    cv = (std_dev / mean_with_zero * 100) if pd.notna(mean_with_zero) and mean_with_zero != 0 else np.nan
    q1 = s_clean.quantile(0.25)
    q2 = s_clean.quantile(0.50)
    q3 = s_clean.quantile(0.75)
    quartiles_str = f"Q1:{q1:.2f}, Q2:{q2:.2f}, Q3:{q3:.2f}"
    
    # Skew and Kurtosis
    skewness = skew(s_clean)
    kurt = kurtosis(s_clean)
    
    # Shortened mean (without bottom 5% and top 5%)
    p5 = s_clean.quantile(0.05)
    p95 = s_clean.quantile(0.95)
    s_trimmed = s_clean[(s_clean >= p5) & (s_clean <= p95)]
    shortened_mean = s_trimmed.mean()

    rows.append({
        'Column': col,
        'Mean (without 0)': round(mean_without_zero, 3) if pd.notna(mean_without_zero) else np.nan,
        'Shortened Mean': round(shortened_mean, 3) if pd.notna(shortened_mean) else np.nan,
        'Variance': round(variance, 3) if pd.notna(variance) else np.nan,
        'Std Dev': round(std_dev, 3) if pd.notna(std_dev) else np.nan,
        'CV (%)': round(cv, 3) if pd.notna(cv) else np.nan,
        'Quartiles': quartiles_str,
        'Skew': round(skewness, 3) if pd.notna(skewness) else np.nan,
        'Kurtosis': round(kurt, 3) if pd.notna(kurt) else np.nan,
        'Zero Count': zero_count,
        'Mean (with 0)': round(mean_with_zero, 3) if pd.notna(mean_with_zero) else np.nan,
        'Mean Change %': round(mean_change_pct, 2) if pd.notna(mean_change_pct) else np.nan,
    })

summary = pd.DataFrame(rows)
print('Blue columns analysis:')
display(summary)

if missing_columns:
    print('\nColumns not found in dataset (check naming):')
    for col in missing_columns:
        print('-', col)
```

```python
from scipy.stats import skew, kurtosis, norm
import math
import textwrap


def classify_distribution(skew_value, kurtosis_value):
    if pd.isna(skew_value) or pd.isna(kurtosis_value):
        return 'Insufficient data'

    if abs(skew_value) < 0.5:
        skew_type = 'approximately symmetric'
    elif skew_value >= 0.5:
        skew_type = 'right-skewed'
    else:
        skew_type = 'left-skewed'

    if abs(kurtosis_value) < 0.5:
        kurtosis_type = 'mesokurtic'
    elif kurtosis_value > 0.5:
        kurtosis_type = 'leptokurtic'
    else:
        kurtosis_type = 'platykurtic'

    if abs(skew_value) < 0.5 and abs(kurtosis_value) < 0.5:
        return f'Approximately normal ({kurtosis_type}, {skew_type})'

    return f'Non-normal ({kurtosis_type}, {skew_type})'


if not resolved_columns:
    print('No matching columns found to visualize.')
else:
    n_cols = 3
    n_rows = math.ceil(len(resolved_columns) / n_cols)
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 5 * n_rows))
    axes = np.array(axes).reshape(-1)

    conclusions = []

    for i, col in enumerate(resolved_columns):
        ax = axes[i]
        s = pd.to_numeric(df[col], errors='coerce').dropna()

        if s.empty:
            ax.set_visible(False)
            conclusions.append({
                'Column': col,
                'Skew': np.nan,
                'Kurtosis': np.nan,
                'Distribution Type': 'Insufficient data'
            })
            continue

        mean_val = s.mean()
        std_val = s.std()
        q1 = s.quantile(0.25)
        q2 = s.quantile(0.50)
        q3 = s.quantile(0.75)

        p5 = s.quantile(0.05)
        p95 = s.quantile(0.95)
        shortened_mean = s[(s >= p5) & (s <= p95)].mean()

        skew_val = skew(s, bias=False)
        kurt_val = kurtosis(s, fisher=True, bias=False)
        dist_type = classify_distribution(skew_val, kurt_val)

        bins = 25
        sns.histplot(s, bins=bins, kde=False, ax=ax, color='#4C78A8', alpha=0.65, edgecolor='white')

        # Overlay fitted normal curve scaled to histogram counts.
        if pd.notna(std_val) and std_val > 0:
            x_vals = np.linspace(s.min(), s.max(), 300)
            bin_width = (s.max() - s.min()) / bins if s.max() > s.min() else 1
            y_vals = norm.pdf(x_vals, loc=mean_val, scale=std_val) * len(s) * bin_width
            ax.plot(x_vals, y_vals, color='navy', linewidth=2.0, label='Normal fit')

        ax.axvline(mean_val, color='red', linestyle='-', linewidth=1.8, label=f'Mean: {mean_val:.2f}')
        ax.axvline(shortened_mean, color='darkorange', linestyle='-', linewidth=1.8, label=f'Short mean: {shortened_mean:.2f}')

        ax.axvline(q1, color='green', linestyle='--', linewidth=1.4, label=f'Q1: {q1:.2f}')
        ax.axvline(q2, color='black', linestyle='--', linewidth=1.4, label=f'Q2: {q2:.2f}')
        ax.axvline(q3, color='purple', linestyle='--', linewidth=1.4, label=f'Q3: {q3:.2f}')

        if pd.notna(std_val):
            ax.axvline(mean_val - std_val, color='brown', linestyle=':', linewidth=1.4, label=f'Mean-Std: {(mean_val - std_val):.2f}')
            ax.axvline(mean_val + std_val, color='brown', linestyle=':', linewidth=1.4, label=f'Mean+Std: {(mean_val + std_val):.2f}')

        stats_text = (
            f"Skew: {skew_val:.3f}\n"
            f"Kurtosis: {kurt_val:.3f}\n"
            f"Type: {dist_type}"
        )
        ax.text(
            0.02, 0.98, stats_text,
            transform=ax.transAxes,
            verticalalignment='top',
            fontsize=9,
            bbox=dict(facecolor='white', alpha=0.8, edgecolor='gray')
        )

        ax.set_title(textwrap.fill(col, width=34), fontsize=11)
        ax.set_xlabel('Value')
        ax.set_ylabel('Frequency')
        ax.legend(fontsize=7, loc='upper right')

        conclusions.append({
            'Column': col,
            'Skew': round(skew_val, 3),
            'Kurtosis': round(kurt_val, 3),
            'Distribution Type': dist_type
        })

    for j in range(len(resolved_columns), len(axes)):
        axes[j].set_visible(False)

    plt.suptitle('Histograms with Normal Fit, Mean, Shortened Mean, Std, Quartiles, Skew, Kurtosis', fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.98])
    plt.show()

    conclusion_df = pd.DataFrame(conclusions)
    print('Distribution conclusion by column:')
    display(conclusion_df)
```

```python
categorical_columns_wanted = [
    'Gender',
    'Customer Type',
    'Type of Travel',
    'Class',
    'Satisfaction',
]

norm_to_actual_cat = {''.join(ch.lower() for ch in str(c) if ch.isalnum()): c for c in df.columns}
resolved_categorical = [
    norm_to_actual_cat[''.join(ch.lower() for ch in col if ch.isalnum())]
    for col in categorical_columns_wanted
    if ''.join(ch.lower() for ch in col if ch.isalnum()) in norm_to_actual_cat
]
missing_categorical = [
    col for col in categorical_columns_wanted
    if ''.join(ch.lower() for ch in col if ch.isalnum()) not in norm_to_actual_cat
]

if not resolved_categorical:
    print('No requested categorical columns found for pie charts.')
else:
    n = len(resolved_categorical)
    n_cols = 3
    n_rows = int(np.ceil(n / n_cols))

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 5 * n_rows))
    axes = np.array(axes).reshape(-1)

    for i, col in enumerate(resolved_categorical):
        ax = axes[i]
        counts = df[col].fillna('Missing').value_counts()

        def autopct_fmt(pct):
            return f'{pct:.1f}%' if pct >= 2 else ''

        ax.pie(
            counts.values,
            labels=counts.index.astype(str),
            autopct=autopct_fmt,
            startangle=90,
            counterclock=False,
            wedgeprops={'edgecolor': 'white', 'linewidth': 1}
        )
        ax.set_title(f'{col} distribution')
        ax.axis('equal')

    for j in range(len(resolved_categorical), len(axes)):
        axes[j].set_visible(False)

    plt.suptitle('Categorical Distributions (Pie Charts)', fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

if missing_categorical:
    print('Columns not found for pie charts:')
    for col in missing_categorical:
        print('-', col)
```
