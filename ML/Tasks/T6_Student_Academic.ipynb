```python
plt.hist(df["final_exam_score"])
plt.title("Final exam score distribution")
plt.xlabel("Final exam score")
plt.ylabel("Count")
plt.show()
```

```python
q1=df["final_exam_score"].quantile(0.25)
q3=df["final_exam_score"].quantile(0.75)
iqr=q3-q1

q2=df["final_exam_score"].quantile(0.5)
price_median=df["final_exam_score"].median()


print(f"Q1:{q1}, Q3:{q3}, IQR:{iqr}")
print(f"{price_median} - {q2}")
```

```python
plt.hist(df["final_exam_score"])
plt.axvline(q1, color="red", linestyle="--")
plt.axvline(q2, color="red", linestyle="--")
plt.axvline(q3, color="red", linestyle="--")
plt.title("Final exam score distribution")
plt.xlabel("Final exam score")
plt.ylabel("Count")
plt.show()
```
